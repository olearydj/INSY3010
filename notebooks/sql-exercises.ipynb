## Setup

````python
import sqlite3
from pathlib import Path
from urllib.request import urlretrieve

db_path = Path("data/flight.db")
if not db_path.exists():
    db_path.parent.mkdir(parents=True, exist_ok=True)
    url = "https://github.com/olearydj/INSY3010/raw/main/notebooks/data/flight.db"
    urlretrieve(url, db_path)
    print("Downloaded flight.db")

conn = sqlite3.connect(db_path)
cur = conn.cursor()


def run(sql):
    """Run a SQL query and print the results."""
    cur.execute(sql)
    cols = [desc[0] for desc in cur.description]
    rows = cur.fetchall()
    widths = [max(len(str(c)), max((len(str(r[i])) for r in rows), default=0)) for i, c in enumerate(cols)]
    header = " | ".join(c.ljust(w) for c, w in zip(cols, widths))
    print(header)
    print("-+-".join("-" * w for w in widths))
    for row in rows:
        print(" | ".join(str(v).ljust(w) for v, w in zip(row, widths)))
    print(f"\n({len(rows)} rows)")
````

### 11. In a single query, find the total number of flights, the earliest departure date, and the latest departure date.

*Concepts: COUNT(\*), MIN, MAX - aggregate functions without GROUP BY*

````python
sql = """
-- your sql here
SELECT
FROM
"""
run(sql)
````

#### Solution

````python
sql = """
SELECT COUNT(*) AS NumFlights,
       MIN(DepartureDate) AS Earliest,
       MAX(DepartureDate) AS Latest
FROM flight
"""
run(sql)
````

### 19. Which passengers have no bookings? Show their email and surname.

*Concepts: LEFT JOIN, WHERE ... IS NULL*

````python
sql = """
-- your sql here
SELECT
FROM
"""
run(sql)
````

#### Solution

````python
sql = """
SELECT passenger.EmailAddress, passenger.Surname
FROM passenger
LEFT JOIN booking
  ON passenger.EmailAddress = booking.EmailAddress
WHERE booking.FlightNumber IS NULL
"""
run(sql)
````

### 20. For each passenger who has at least one booking, how many flights have they booked? Label the count `NumFlights` and sort by number of flights descending, then by surname.

*Concepts: INNER JOIN, GROUP BY, COUNT(\*), HAVING, ORDER BY, AS*

````python
sql = """
-- your sql here
SELECT
FROM
"""
run(sql)
````

#### Solution

````python
sql = """
SELECT passenger.Surname, passenger.GivenNames,
       COUNT(*) AS NumFlights
FROM passenger
INNER JOIN booking
  ON passenger.EmailAddress = booking.EmailAddress
GROUP BY passenger.EmailAddress, passenger.Surname, passenger.GivenNames
HAVING COUNT(*) >= 1
ORDER BY NumFlights DESC, passenger.Surname
"""
run(sql)
````
