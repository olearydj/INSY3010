Write a function `is_palindrome` that determines if a string is a palindrome. A palindrome is text like `racecar` that reads the same forward and backward, ignoring spaces, punctuation, and case. Return a boolean value.

You do not need to handle numeric values or empty strings. A single character is a palindrome.

````python
# define your function below
````

````python
# test cases
# print(is_palindrome("racecar"))                          # True
# print(is_palindrome("nope"))                             # False
# print(is_palindrome("A man, a plan, a canal, Panama"))   # True
# print(is_palindrome("a"))                                # True
# print(is_palindrome("A Toyota! A Toyota!"))              # True
````

#### Solution

````python
def is_palindrome(text):
    cleaned = ""
    for c in text:
        if c.isalpha():
            cleaned += c.lower()
    return cleaned == cleaned[::-1]


print(is_palindrome("racecar"))                          # True
print(is_palindrome("nope"))                             # False
print(is_palindrome("A man, a plan, a canal, Panama"))   # True
print(is_palindrome("a"))                                # True
print(is_palindrome("A Toyota! A Toyota!"))              # True
````

Notes:

- Two moves: clean the text (keep only letters, lowercase them), then compare it to its reverse. Both are well within Test 1 material.
- `c.isalpha()` is `True` for letters and `False` for spaces and punctuation - exactly what we want to filter.
- `cleaned[::-1]` reverses a string. Same slice notation you've used on lists.

## 3. Inventory Value

Write a function `calculate_inventory_value` that takes two dictionaries:

1. `inventory`, with `part_number: quantity` pairs.
2. `prices`, with `part_number: price` pairs.

Return the total value of the inventory - the sum of `quantity * price` for every part. If a part has no corresponding price, treat the price as `0`.

````python
# define your function below
````

````python
# test
inventory = {"A": 1, "B": 2, "C": 3, "D": 1000}
prices = {"A": 10, "B": 20, "C": 30}
# print(calculate_inventory_value(inventory, prices))  # 140
````

#### Solution

````python
def calculate_inventory_value(inventory, prices):
    total = 0
    for part, qty in inventory.items():
        price = prices.get(part, 0)
        total += qty * price
    return total


inventory = {"A": 1, "B": 2, "C": 3, "D": 1000}
prices = {"A": 10, "B": 20, "C": 30}
print(calculate_inventory_value(inventory, prices))  # 140
````

Notes:

- Loop over `inventory.items()` to get each `(part, qty)` pair.
- `prices.get(part, 0)` returns the price if `part` is a key, otherwise `0`. This is exactly the "if an item has no price, use zero" rule, expressed in one call instead of an `if` check.
- Accumulator pattern: initialize `total = 0`, add to it inside the loop, return it at the end.

## 4. Deg_C Analysis

Write a function `analyze_temperatures` that takes a list of temperatures in Fahrenheit and returns a tuple containing the minimum, maximum, mean, and standard deviation of the temperatures in *Celsius*. Use the supplied `stdev` function for the standard deviation, and round that value to 2 decimal places. `analyze_temperatures` must call a second function, `convert_to_celsius`, to convert a single Fahrenheit value. Formula: `C = (F - 32) * 5/9`.

For `[32, 68, 86, 104, 212]` the expected result is `(0.0, 100.0, 38.0, 37.68)`.

````python
# do not modify the provided stdev function
def stdev(vals):
    """Sample standard deviation for a sequence of values."""
    if len(vals) == 0:
        return None
    if len(vals) == 1:
        return 0

    mean = sum(vals) / len(vals)
    squared_diffs = [(val - mean) ** 2 for val in vals]
    variance = sum(squared_diffs) / (len(vals) - 1)
    return variance ** 0.5


# define analyze_temperatures and convert_to_celsius below
````

````python
# test
f_temps = [32, 68, 86, 104, 212]
# print(analyze_temperatures(f_temps))  # (0.0, 100.0, 38.0, 37.68)
````

#### Solution

````python
def convert_to_celsius(f):
    return (f - 32) * 5 / 9


def analyze_temperatures(f_temps):
    c_temps = []
    for f in f_temps:
        c_temps.append(convert_to_celsius(f))

    return (
        min(c_temps),
        max(c_temps),
        sum(c_temps) / len(c_temps),
        round(stdev(c_temps), 2),
    )


f_temps = [32, 68, 86, 104, 212]
print(analyze_temperatures(f_temps))  # (0.0, 100.0, 38.0, 37.68)
````

Notes:

- Two functions working together. `convert_to_celsius` handles one value; `analyze_temperatures` handles the whole list and calls the first for each element.
- Standard accumulator loop to build the Celsius list. A list comprehension (`[convert_to_celsius(f) for f in f_temps]`) would also work if you are comfortable with them.
- `min`, `max`, `sum`, `len` are all built-ins that take an iterable. Four summary values, all expressed directly.
- The returned tuple has four elements. Wrapping them in parentheses is optional (Python auto-packs comma-separated values into a tuple) but reads clearer with them.

````python
# test
csv_data = [
    ["id", "name", "category", "price"],
    ["101", "Laptop", "Electronics", "999.99"],
    ["102", "Desk Chair", "Furniture", "189.50"],
    ["103", "Coffee Maker", "Appliances", "49.99"],
    ["104", "Smartphone", "Electronics", "599.99"],
    ["105", "Dining Table", "Furniture", "349.99"],
    ["106", "USB Cable", "Electronics", "9.99"],
]
# print(categorize_products(csv_data))
# (["Coffee Maker", "USB Cable"], ["Desk Chair", "Dining Table"], ["Laptop", "Smartphone"])
````

#### Solution

````python
def categorize_products(csv_data):
    budget = []
    mid_range = []
    premium = []

    for row in csv_data[1:]:          # skip the header row
        name = row[1]
        price = float(row[3])         # price is a string in CSV data

        if price < 100:
            budget.append(name)
        elif price < 500:
            mid_range.append(name)
        else:
            premium.append(name)

    return (budget, mid_range, premium)


csv_data = [
    ["id", "name", "category", "price"],
    ["101", "Laptop", "Electronics", "999.99"],
    ["102", "Desk Chair", "Furniture", "189.50"],
    ["103", "Coffee Maker", "Appliances", "49.99"],
    ["104", "Smartphone", "Electronics", "599.99"],
    ["105", "Dining Table", "Furniture", "349.99"],
    ["106", "USB Cable", "Electronics", "9.99"],
]
print(categorize_products(csv_data))
````

Notes:

- Three things going on in one function: skip the header, parse each row, sort into the right bucket. Each piece is simple; the care is in combining them.
- `csv_data[1:]` drops the header row. `row[1]` is the name column, `row[3]` is the price column. Remember CSV rows come in as strings - `float(row[3])` converts before comparing.
- The `if` / `elif` / `else` chain sorts into buckets. The order matters: check smaller ranges first so the `elif` only catches values that failed the earlier test.

Write a function `read_csv_rows(path)` that opens a CSV file and returns every row as a list of lists (one inner list per row, including the header). If any error occurs while opening or reading, catch it and return an empty list `[]` instead. Use a `with` statement to open the file.

````python
# define your function below
````

````python
# test
# rows = read_csv_rows("data/5k-sales.csv")
# print(len(rows))        # 5001 (1 header + 5000 data rows)
# print(rows[0][:3])      # ['Order_ID', 'Region', 'Country']
# print(read_csv_rows("does_not_exist.csv"))  # []
````

#### Solution

````python
import csv


def read_csv_rows(path):
    try:
        with open(path) as f:
            reader = csv.reader(f)
            rows = []
            for row in reader:
                rows.append(row)
            return rows
    except:
        return []


rows = read_csv_rows("data/5k-sales.csv")
print(len(rows))        # 5001
print(rows[0][:3])      # ['Order_ID', 'Region', 'Country']
print(read_csv_rows("does_not_exist.csv"))  # []
````

Notes:

- `with open(path) as f:` opens the file and guarantees it is closed when the block ends, even if an error occurs inside.
- `csv.reader(f)` wraps the file handle; iterating over it yields one list per row.
- The bare `except:` block catches any error that happens inside the `try` - file missing, permission denied, malformed data, whatever - and returns `[]` instead of letting it crash the program. For production code, catching a specific exception (like `FileNotFoundError`) is usually better because it only suppresses the error you expect.

## 7. Count Rows in a SQLite Database

Write a function `count_orders(db_path, region)` that returns the number of rows in the `sales` table of a SQLite database where `Region` equals the given value. Use a parameterized query to pass the region safely.

````python
# define your function below
````

````python
# test
# print(count_orders("data/5k-sales.db", "Asia"))       # 719
# print(count_orders("data/5k-sales.db", "Europe"))     # 1330
# print(count_orders("data/5k-sales.db", "Atlantis"))   # 0
````

#### Solution

````python
import sqlite3


def count_orders(db_path, region):
    with sqlite3.connect(db_path) as con:
        cur = con.cursor()
        cur.execute(
            "SELECT COUNT(*) FROM sales WHERE Region = ?",
            (region,),
        )
        return cur.fetchone()[0]


print(count_orders("data/5k-sales.db", "Asia"))       # 719
print(count_orders("data/5k-sales.db", "Europe"))     # 1330
print(count_orders("data/5k-sales.db", "Atlantis"))   # 0
````

Notes:

- `with sqlite3.connect(db_path) as con:` is the same pattern as `with open`: it manages the connection and closes it when the block exits.
- The `?` placeholder in the SQL string is replaced by the value in the tuple passed as the second argument to `execute`. The trailing comma in `(region,)` is what makes it a one-element tuple, not just parentheses. Never use f-strings or string concatenation to insert user input into SQL - that is what makes queries vulnerable to injection.
- `COUNT(*)` returns a single row with one column, so `fetchone()` gives you that row and `[0]` pulls out the count value. A nonexistent region still returns one row; the count is just `0`.
