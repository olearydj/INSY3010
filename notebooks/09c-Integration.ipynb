## Sorting and Ordering

In the Dictionaries notebook we briefly saw two ways to sort data: the built-in `sorted` function and the `.sort` method. Let's explore them fully, including options for controlling sort order and sorting by custom criteria.

### `sorted()` vs `.sort()` - Review

Recall the key distinction: `sorted()` returns a *new* sorted list, while `.sort()` sorts a list *in place* and returns `None`.

````python
numbers = [3, 1, 4, 1, 5, 9, 2, 6]

# sorted() returns a new list - original unchanged
result = sorted(numbers)
print("sorted():", result)
print("original:", numbers)
````

````python
# .sort() modifies the list in place - returns None
return_value = numbers.sort()
print("after .sort():", numbers)
print("return value:", return_value)
````

A common mistake is trying to capture the return value of `.sort()`:

````python
data = [5, 3, 8, 1]
data = data.sort()  # data is now None!
print(data)
````

This is the same pattern we saw with `.append()` - in-place methods return `None`, not the modified object.

Another important distinction: `sorted()` works on *any* iterable (strings, tuples, sets, dictionaries), but always returns a list. The `.sort()` method is only available on lists.

````python
# sorted() works on any iterable
print(sorted("python"))
print(sorted((9, 3, 7, 1)))
print(sorted({5, 2, 8}))
````

### The `key` Parameter

Both `sorted()` and `.sort()` accept a `key` parameter that controls *what* values are compared during sorting. The `key` must be a function - Python calls it on each element and sorts by the returned values.

The simplest use is with a built-in function. For example, to sort strings by length instead of alphabetically:

````python
words = ["banana", "kiwi", "strawberry", "fig", "apple"]

# default: alphabetical order
print(sorted(words))

# sort by string length
print(sorted(words, key=len))
````

Note that `key=len` passes the function itself - no parentheses. Python calls `len` on each word internally to determine the sort order.

Another useful built-in is `str.lower`, which enables case-insensitive sorting:

````python
names = ["Charlie", "alice", "Bob", "dana"]

# default: uppercase letters sort before lowercase
print(sorted(names))

# case-insensitive: compare lowercase versions
print(sorted(names, key=str.lower))
````

### Sorting with Custom Functions

When built-in functions aren't sufficient, you can write your own. This is especially useful for sorting complex data structures like lists of tuples.

Consider a list of student records where each record is a `(name, score)` tuple:

````python
students = [
    ("Charlie", 85),
    ("Alice", 92),
    ("Bob", 78),
    ("Dana", 92),
    ("Eli", 88)
]

# default sort: alphabetical by first element (name)
print(sorted(students))
````

To sort by score instead, write a helper function that extracts the value to sort by:

````python
def by_score(student):
    return student[1]

sorted(students, key=by_score)
````

The function must accept a single argument (one element from the list) and return the value to compare. Python calls it on every element, then sorts based on those returned values.

### The `reverse` Parameter

Both `sorted()` and `.sort()` accept a `reverse` parameter. When set to `True`, the result is in descending order.

````python
numbers = [3, 1, 4, 1, 5, 9, 2, 6]

print(sorted(numbers))                 # ascending (default)
print(sorted(numbers, reverse=True))   # descending
````

You can combine `key` and `reverse`:

````python
# sort students by score, highest first
sorted(students, key=by_score, reverse=True)
````

### `reversed()` vs `.reverse()`

Python also provides two ways to reverse (not sort) a sequence. The pattern mirrors `sorted()` vs `.sort()`:

- `.reverse()` reverses a list *in place* and returns `None`
- `reversed()` returns an *iterator* over the elements in reverse order

````python
letters = ['a', 'b', 'c', 'd']

# .reverse() modifies in place, returns None
return_value = letters.reverse()
print("after .reverse():", letters)
print("return value:", return_value)
````

````python
letters = ['a', 'b', 'c', 'd']

# reversed() returns an iterator - wrap in list() to see the result
rev = reversed(letters)
print("iterator:", rev)
print("as list:", list(rev))
print("original:", letters)  # unchanged
````

The iterator returned by `reversed()` behaves like `range()` - it produces values on demand but isn't a list itself. Wrap it in `list()` when you need to see or store the result.

### Summary Table

| | New object | In place |
|:---|:---|:---|
| Sort | `sorted(seq)` → new list | `list.sort()` → `None` |
| Reverse | `reversed(seq)` → iterator | `list.reverse()` → `None` |

Both `sorted()` and `.sort()` accept `key=` and `reverse=` parameters.
`reversed()` and `.reverse()` simply flip the existing order without comparing values.

## Nested Data Structures

So far, our lists, tuples, and dictionaries have contained simple values like numbers and strings. But containers can hold *other containers* as their elements. This is called *nesting*, and it's how Python represents structured, multi-dimensional data like tables, records, and hierarchies.

### Nested Lists and Tuples

Much of the world's data is found in tables. The tabular format - rows, columns, and cells - is most commonly represented in Python using *nested* lists or tuples: lists that contain lists as elements.

````python
list_of_lists = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
````

While valid, this is easier to read when formatted to show the table structure:

````python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
````

Inner lists represent rows; elements within each row correspond to columns. Python ignores whitespace inside brackets, so this formatting is purely for readability.

### Accessing Nested Sequences

To access an element in a nested list, start by getting the desired row, then index into it:

````python
row = matrix[1]       # second row: [4, 5, 6]
element = row[1]      # second element: 5
print(element)
````

These two steps can be combined into a single expression:

````python
print(matrix[1][1])   # 5
````

Think of each `[]` as one indexing operation, evaluated left to right. `matrix[1]` gives us the list `[4, 5, 6]`, then `[1]` gives us the element `5`.

This is the same pattern you've already used to extract characters from strings in lists:

````python
words = ["list", "of", "strings"]
print(words[-1][-1])  # last character of last word: 's'
````

### Modifying Nested Lists

Since lists are mutable, you can assign to elements within nested lists using the same double-index syntax:

````python
matrix[2][2] = 42
print(matrix)
````

Because `matrix[2]` is itself a list, any list operation works:

````python
matrix[2].append(10)
print(matrix)
````

````python
# reset for later examples
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
````

### Nested Loops for Traversal

Nested loops naturally match the structure of nested data. When you need to visit every element in a nested list, use one loop for the rows and another for the elements within each row.

Here's a tic-tac-toe board stored as a list of lists:

````python
board = [
    ['X', 'O', 'X'],
    ['O', 'X', 'O'],
    ['O', 'X', 'X']
]

def show_board(board):
    '''display the board for the player'''
    for row in board:
        for cell in row:
            print(cell, end=' ')
        print()  # newline after each row

show_board(board)
````

The outer loop handles the rows, while the inner loop handles the cells within each row.

### Nested Loops for Creation / Modification

After learning to traverse a nested list, a natural next step is to build one. Here's a function that multiplies every element of a matrix by a scalar:

````python
def scalar_multiply(matrix, scalar):
    result = []
    for row in matrix:
        new_row = []
        for element in row:
            new_row.append(element * scalar)
        result.append(new_row)
    return result

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

scalar_multiply(matrix, 2)
````

The function builds a new nested list rather than modifying the original - a good practice when you want to preserve the input data.

### Nested Dictionaries

Dictionary values can themselves be dictionaries. This is useful for representing structured records where each entry has multiple attributes.

The following example models a parts inventory. Top-level keys are part numbers; each maps to a dictionary of details about that part.

````python
parts_inventory = {
    "A101": {
        "name": "Gear",
        "quantity": 50,
        "unit_cost": 15.00,
        "supplier": "GearCo"
    },
    "B202": {
        "name": "Bearing",
        "quantity": 100,
        "unit_cost": 5.50,
        "supplier": "BearingInc"
    },
    "C303": {
        "name": "Belt",
        "quantity": 75,
        "unit_cost": 8.25,
        "supplier": "BeltMasters"
    }
}
````

Nested lookups access the inner values:

````python
# quantity of Gears in stock
print(parts_inventory["A101"]["quantity"])
````

````python
# update the unit cost of Bearings
parts_inventory["B202"]["unit_cost"] = 5.75
print(parts_inventory["B202"])
````

````python
# add a new part
parts_inventory["D404"] = {
    "name": "Pulley",
    "quantity": 30,
    "unit_cost": 12.50,
    "supplier": "PulleyCorp"
}
````

### Dictionaries with List Values

A common pattern is dictionaries whose values are lists. This lets you accumulate multiple items under each key.

````python
student_grades = {
    "Jack": [89, 78, 92],
    "Jill": [82, 98, 79],
    "John": [42, 67, 55]
}

# access Jack's grades
print(student_grades["Jack"])

# add a new grade for Jack
student_grades["Jack"].append(100)
print(student_grades["Jack"])
````

Remember from the Dictionaries notebook that aliasing applies here. If two keys point to the *same* list object, modifying through one key affects the other. When in doubt, create independent copies with `list()`.

Write a function `calculate_total_inventory_value` that takes a dictionary in the same structure as `parts_inventory` and returns the total value of all parts (quantity times unit cost, summed across all parts).

#### Solution / Discussion

````python
def calculate_total_inventory_value(parts):
    total = 0
    for part_data in parts.values():
        total += part_data["quantity"] * part_data["unit_cost"]
    return total

calculate_total_inventory_value(parts_inventory)
````

Looping over `.values()` gives direct access to each inner dictionary, avoiding the extra lookup needed when looping over keys. Compare this to the alternative:

````python
def calculate_total_inventory_value_v2(parts):
    total = 0
    for part_id in parts:
        total += parts[part_id]["quantity"] * parts[part_id]["unit_cost"]
    return total

calculate_total_inventory_value_v2(parts_inventory)
````

Both work, but the `.values()` version is cleaner when you don't need the keys.
